```python
"""
Binance multi-instrument, multi-alpha, multi-timeframe bot (testnet-ready)

Features:
- Subscribe to combined kline websockets (testnet) for multiple symbols & timeframes
- Maintain rolling buffer (last N candles, default 40) per symbol/timeframe
- Call multiple alpha functions (user-supplied) on candle data; alphas return -1/0/1
- Simple position manager: place market buy/sell on testnet via python-binance
- Persist candles and order events to an SQLite DB
- Reconnect logic and basic rate limiting

Usage:
- Edit CONFIG at top: API_KEY, API_SECRET, SYMBOLS, TIMEFRAMES, ALPHAS
- Run: python binance_multialpha_bot.py

Requirements:
- python-binance
- websockets
- asyncio
- pandas
- sqlite3 (std lib)

This file is intended as a practical, extensible starting point. Replace `alpha_x`
with your real alpha functions; they should accept (symbol, timeframe, candles_df)
and return -1/0/1.
"""

import asyncio
import json
import time
import sqlite3
from collections import deque, defaultdict
from datetime import datetime
import logging

import pandas as pd
import websockets
from binance.client import Client

# ----------------- CONFIG -----------------
# API_KEY = "<YOUR_TESTNET_API_KEY>"
# API_SECRET = "<YOUR_TESTNET_API_SECRET>"
# use testnet REST base
REST_BASE = "https://testnet.binance.vision"
# websocket base for spot testnet (official doc)
WS_BASE = "wss://stream.testnet.binance.vision/stream?streams="

# instruments and timeframes
SYMBOLS = ["BTCUSDT", "SOLUSDT"]  # add more uppercase symbols
TIMEFRAMES = ["1m"]  # you can add '5m','15m' etc
CANDLES_TO_KEEP = 40

# Alphas: list of callables. Replace with your own implementations.
# Each alpha should return: 1 (buy), 0 (no-op), -1 (sell/short)

def alpha_momentum(symbol, timeframe, candles_df):
    """Simple momentum: buy if last close > mean of last 10 closes, sell if below."""
    if len(candles_df) < 11:
        return 0
    last = candles_df['Close'].iloc[-1]
    mean10 = candles_df['Close'].iloc[-11:-1].mean()
    if last > mean10 * 1.001:
        return 1
    if last < mean10 * 0.999:
        return -1
    return 0


def alpha_mean_revert(symbol, timeframe, candles_df):
    """Simple mean-reversion: if price spikes above short MA, signal sell."""
    if len(candles_df) < 21:
        return 0
    short = candles_df['Close'].rolling(5).mean().iloc[-1]
    long = candles_df['Close'].rolling(20).mean().iloc[-1]
    if short > long * 1.002:
        return -1
    if short < long * 0.998:
        return 1
    return 0

ALPHAS = [alpha_momentum, alpha_mean_revert]
# ------------------------------------------

logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s:%(message)s")

# sqlite helper
DB_FILE = "bot_data.db"

def init_db():
    conn = sqlite3.connect(DB_FILE)
    c = conn.cursor()
    c.execute('''
        CREATE TABLE IF NOT EXISTS candles (
            symbol TEXT, timeframe TEXT, open_time INTEGER, open REAL, high REAL, low REAL, close REAL, volume REAL, close_time INTEGER
        )
    ''')
    c.execute('''
        CREATE TABLE IF NOT EXISTS orders (
            id TEXT PRIMARY KEY, symbol TEXT, side TEXT, qty REAL, price REAL, status TEXT, timestamp INTEGER, raw TEXT
        )
    ''')
    conn.commit()
    conn.close()

# data buffers: buffers[symbol][timeframe] = deque of dicts
buffers = defaultdict(lambda: defaultdict(lambda: deque(maxlen=CANDLES_TO_KEEP)))

# position manager: naive: store current qty per symbol
positions = defaultdict(float)  # positive -> long qty, negative -> short qty

# REST client
client = Client(API_KEY, API_SECRET)
client.API_URL = REST_BASE

# small helper: convert websocket kline payload to dict

def kline_payload_to_row(k):
    return {
        'open_time': int(k['t']),
        'open': float(k['o']),
        'high': float(k['h']),
        'low': float(k['l']),
        'close': float(k['c']),
        'volume': float(k['v']),
        'close_time': int(k['T'])
    }

async def persist_candle(symbol, timeframe, row):
    conn = sqlite3.connect(DB_FILE)
    c = conn.cursor()
    c.execute('''INSERT INTO candles(symbol,timeframe,open_time,open,high,low,close,volume,close_time)
                 VALUES (?,?,?,?,?,?,?,?,?)''', (
        symbol, timeframe, row['open_time'], row['open'], row['high'], row['low'], row['close'], row['volume'], row['close_time']
    ))
    conn.commit()
    conn.close()

async def persist_order(resp):
    conn = sqlite3.connect(DB_FILE)
    c = conn.cursor()
    order_id = str(resp.get('orderId', str(time.time())))
    c.execute('''INSERT OR REPLACE INTO orders(id,symbol,side,qty,price,status,timestamp,raw)
                 VALUES (?,?,?,?,?,?,?,?)''', (
        order_id,
        resp.get('symbol'),
        resp.get('side'),
        float(resp.get('origQty', 0)),
        float(resp.get('price', 0)),
        resp.get('status', ''),
        int(time.time()*1000),
        json.dumps(resp)
    ))
    conn.commit()
    conn.close()

async def handle_signal(symbol, timeframe, signal):
    """Very simple execution logic:
       - if signal == 1 and not holding -> place market buy
       - if signal == -1 and holding -> market sell to flat
       - more sophisticated sizing/management can be added
    """
    current_qty = positions.get(symbol, 0)
    logging.info(f"Signal for {symbol} {timeframe}: {signal}, current_qty={current_qty}")

    if signal == 1 and current_qty == 0:
        # place a small market buy (testnet only)
        qty = 0.001 if symbol.startswith('BTC') else 1
        try:
            resp = client.order_market_buy(symbol=symbol, quantity=qty)
            positions[symbol] += float(qty)
            logging.info(f"Bought {qty} {symbol}")
            await persist_order(resp)
        except Exception as e:
            logging.exception("Order failed")

    elif signal == -1 and current_qty > 0:
        # sell all
        qty = current_qty
        try:
            resp = client.order_market_sell(symbol=symbol, quantity=qty)
            positions[symbol] -= float(qty)
            logging.info(f"Sold {qty} {symbol}")
            await persist_order(resp)
        except Exception as e:
            logging.exception("Sell failed")

# construct combined stream name list

def make_streams(symbols, timeframes):
    streams = []
    for s in symbols:
        for tf in timeframes:
            streams.append(f"{s.lower()}@kline_{tf}")
    return "/".join(streams)

async def ws_loop(symbols, timeframes):
    streams = make_streams(symbols, timeframes)
    url = WS_BASE + streams
    logging.info(f"Connecting to {url}")

    while True:
        try:
            async with websockets.connect(url, ping_interval=60) as ws:
                logging.info("Websocket connected")
                async for message in ws:
                    data = json.loads(message)
                    # combined stream wraps {"stream":..., "data":...}
                    payload = data.get('data') if 'data' in data else data
                    if payload is None:
                        continue
                    if payload.get('e') == 'kline' or payload.get('e') == 'kline':
                        k = payload['k']
                        symbol = payload['s']
                        timeframe = k['i']
                        row = kline_payload_to_row(k)

                        # store in buffer
                        buffers[symbol][timeframe].append(row)

                        # persist
                        await persist_candle(symbol, timeframe, row)

                        # prepare pandas df for alphas
                        df = pd.DataFrame(list(buffers[symbol][timeframe]))
                        df = df.sort_values('open_time')
                        df['Close'] = df['close']

                        # evaluate alphas
                        for alpha in ALPHAS:
                            try:
                                sig = alpha(symbol, timeframe, df)
                                if sig != 0:
                                    await handle_signal(symbol, timeframe, sig)
                            except Exception:
                                logging.exception(f"Alpha {alpha.__name__} failed for {symbol} {timeframe}")

        except Exception as e:
            logging.exception("Websocket connection failed - retrying in 5s")
            await asyncio.sleep(5)

# helper task: poll account and store positions
async def account_poller():
    while True:
        try:
            acc = client.get_account()
            for b in acc['balances']:
                free = float(b['free'])
                if free > 0:
                    positions[b['asset']+'USDT'] = positions.get(b['asset']+'USDT', 0)  # naive mapping
            logging.debug("Account polled")
        except Exception:
            logging.exception("Account poll failed")
        await asyncio.sleep(30)

async def main():
    init_db()
    # pre-fill buffers with REST klines for last CANDLES_TO_KEEP
    for s in SYMBOLS:
        for tf in TIMEFRAMES:
            kl = client.get_klines(symbol=s, interval=tf, limit=CANDLES_TO_KEEP)
            for k in kl:
                row = {
                    'open_time': int(k[0]),
                    'open': float(k[1]),
                    'high': float(k[2]),
                    'low': float(k[3]),
                    'close': float(k[4]),
                    'volume': float(k[5]),
                    'close_time': int(k[6])
                }
                buffers[s][tf].append(row)

    # start websocket listener + account poller
    await asyncio.gather(ws_loop(SYMBOLS, TIMEFRAMES), account_poller())

if __name__ == '__main__':
    try:
        asyncio.run(main())
    except KeyboardInterrupt:
        logging.info('Shutting down')
```
